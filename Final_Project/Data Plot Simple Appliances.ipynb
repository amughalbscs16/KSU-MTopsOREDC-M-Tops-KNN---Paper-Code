```python
import csv
import matplotlib.pyplot as plt 
with open('appliances.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    data = []
    Airconditioner = []
    Kettle = []
    Iron = []
    WaterDispensor = []
    for row in csv_reader:
        data.append(row)
    #removing headers
    data = data[2:]
    
    Airconditioner = [float(elem[0]) for elem in data]
    timesteps = [i for i in range(len(data))]
    Kettle = [float(elem[1]) for elem in data]
    Iron = [float(elem[2]) for elem in data]
    WaterDispensor = [float(elem[3]) for elem in data]
    #Air Conditioner
    plt.plot(timesteps,Airconditioner)
  
    # naming the x axis 
    plt.xlabel('timestep')
    # naming the y axis 
    plt.ylabel('current') 

    # giving a title to my graph 
    plt.title('Current-Airconditioner') 

    # function to show the plot 
    plt.show() 
    #Kettle
    plt.plot(timesteps,Kettle)
  
    # naming the x axis 
    plt.xlabel('timestep')
    # naming the y axis 
    plt.ylabel('current') 

    # giving a title to my graph 
    plt.title('Current-Kettle') 

    # function to show the plot 
    plt.show() 
    plt.plot(timesteps,Iron)
  
    # naming the x axis 
    plt.xlabel('timestep')
    # naming the y axis 
    plt.ylabel('current') 

    # giving a title to my graph 
    plt.title('Current-Iron') 

    # function to show the plot 
    plt.show() 
    plt.plot(timesteps,WaterDispensor)
  
    # naming the x axis 
    plt.xlabel('timestep')
    # naming the y axis 
    plt.ylabel('current') 

    # giving a title to my graph 
    plt.title('Current-WaterDispensor') 

    # function to show the plot 
    plt.show() 
    
    #print(timesteps,Airconditioner)
```
